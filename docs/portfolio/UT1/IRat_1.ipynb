* **Definici√≥n**: muestran la **distribuci√≥n** de una variable continua.

# **Lectura: 2 "Buen an√°lisis de datos"**

El **an√°lisis de datos** es poderoso pero propenso a errores.
Un buen analista se distingue por ser **cuidadoso y met√≥dico**, lo que genera **credibilidad** en sus conclusiones.
Este documento resume t√©cnicas y mentalidades necesarias para trabajar con **grandes conjuntos de datos de alta dimensi√≥n**, como los que maneja Google.

## üîπ 1. Aspectos t√©cnicos

### **Consultar distribuciones**

* No basta con **media, mediana o desviaci√≥n est√°ndar**.
* Hay que mirar **histogramas, CDF, Q-Q plots**, etc.
* Permiten detectar **multimodalidad** o **valores at√≠picos**.

### **Valores at√≠picos**

* Son se√±ales de alerta, pueden indicar **errores en el proceso o fen√≥menos reales**.
* Pueden excluirse, pero siempre entendiendo su origen.
* Ejemplo: consultas con muy pocos clics pueden se√±alar problemas de registro.

### **Ruido**

* El **ruido aleatorio no desaparece** aunque haya muchos datos.
* Usar **intervalos de confianza, p-values** para medir la certeza.

### **Consultar ejemplos**

* Revisar **muestras individuales** para validar el an√°lisis.
* Mirar casos **extremos** (5% m√°s r√°pido/lento) y todas las clases.
* Garantiza que los res√∫menes no oculten errores.

### **Dividir los datos (segmentaci√≥n)**

* Separar por **navegador, dispositivo, regi√≥n, dominio**, etc.
* Detecta incoherencias internas.
* Cuidado con el **cambio de mezcla** ‚Üí riesgo de la **Paradoja de Simpson**.

### **Importancia pr√°ctica**

* Diferenciar entre **significancia estad√≠stica** y **relevancia real**.
* Preguntarse: ¬øun cambio de 0.1% realmente importa?

### **Coherencia a lo largo del tiempo**

* Analizar datos por **unidades temporales** (d√≠a, mes).
* Detectar anomal√≠as y verificar su causa antes de descartarlas.
* Sirve para validar variabilidad e importancia estad√≠stica.

### **Filtros y proporciones**

* Siempre declarar los **filtros aplicados** y contar cu√°ntos datos se excluyen.
* Aclarar numerador y denominador en **proporciones**.

  * Ejemplo: ‚ÄúB√∫squedas por usuario‚Äù puede tener distintas definiciones ‚Üí hay que ser expl√≠cito.


## üîπ 2. Procesos

### **Tres etapas del an√°lisis**

1. **Validaci√≥n** ‚Üí ¬ølos datos son correctos, coherentes y representan lo esperado?
2. **Descripci√≥n** ‚Üí interpretaci√≥n objetiva (qu√© muestran los datos).
3. **Evaluaci√≥n** ‚Üí juicio de valor (¬øes bueno o malo para usuario, empresa, sociedad?).

Separar estas etapas evita **sesgos y malas interpretaciones**.

### **Confirmar configuraci√≥n de experimentos**

* Revisar **c√≥mo y cu√°ndo** se recopilaron los datos.
* Detectar sesgos (ej: solo usuarios de Chrome, per√≠odos festivos, etc.).

### **M√©tricas est√°ndar vs. personalizadas**

* Siempre empezar con **m√©tricas est√°ndar** (clics, tiempo de carga, errores).
* Luego pasar a m√©tricas nuevas ‚Üí validarlas comparando con las est√°ndar.

### **Medir varias veces**

* Capturar un mismo fen√≥meno de distintas formas.
* Usar fuentes de datos distintas para confirmar.

### **Reproducibilidad**

* Un hallazgo debe mantenerse en **diferentes poblaciones, tiempos y muestras aleatorias**.
* Si no, probablemente es un error.

### **Comparar con mediciones anteriores**

* Validar con resultados previos.
* Un n√∫mero muy diferente debe ser tratado como **sospechoso** hasta validarlo.

### **Aplicar m√©tricas nuevas en datos viejos**

* Antes de confiar en una m√©trica nueva, probarla en datos conocidos.
* Ejemplo: una m√©trica de satisfacci√≥n deber√≠a reflejar funciones que ya sabemos que son buenas.

### **Hip√≥tesis y evidencia**

* No basta con generar teor√≠as ‚Üí hay que **buscar evidencia que las valide o refute**.
* Preguntarse: ¬øqu√© experimento podr√≠a confirmar esto?
* Esto fomenta descubrimientos y m√©tricas nuevas.

### **Iteraci√≥n r√°pida en an√°lisis exploratorio**

* Hacer ciclos completos r√°pido, no buscar perfecci√≥n en el primer paso.
* Iterar permite descubrir problemas antes.

### **Cuidado con retroalimentaci√≥n (feedback loops)**

* Si la m√©trica de √©xito alimenta el sistema, no puede usarse luego como criterio de evaluaci√≥n.
* Ejemplo: si mostramos m√°s anuncios ‚Üí m√°s clics no implica autom√°ticamente m√°s satisfacci√≥n.


## üîπ 3. Mentalidad

### **Comenzar con preguntas, no con datos**

* El an√°lisis siempre debe partir de **hip√≥tesis o preguntas claras**.
* Evita caer en usar ‚Äútu t√©cnica favorita‚Äù sin necesidad real.

### **Ser esc√©ptico y defensor**

* Un indicador puede ser un **proxy** de otro, pero siempre hay que testearlo.

### **Compartir primero con pares**

* Revisiones entre colegas son clave para detectar errores y sesgos.

### **Aceptar errores e ignorancia**

## ‚úÖ Reflexiones finales
